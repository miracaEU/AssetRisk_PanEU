```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Define countries and infrastructure types
COUNTRIES = ['AUT', 'BEL', 'BGR', 'CHE', 'CYP', 'CZE', 'DEU', 'DNK', 'ESP', 'EST', 
             'FIN', 'FRA', 'GRC', 'HRV', 'HUN', 'IRL', 'ISL', 'ITA', 'LIE', 'LTU', 
             'LUX', 'LVA', 'MLT', 'NLD', 'NOR', 'POL', 'PRT', 'ROU', 'SVK', 'SVN', 'SWE']

# Country code to full name mapping
COUNTRY_NAMES = {
    'AUT': 'Austria',
    'BEL': 'Belgium', 
    'BGR': 'Bulgaria',
    'CHE': 'Switzerland',
    'CYP': 'Cyprus',
    'CZE': 'Czech Republic',
    'DEU': 'Germany',
    'DNK': 'Denmark',
    'ESP': 'Spain',
    'EST': 'Estonia',
    'FIN': 'Finland',
    'FRA': 'France',
    'GRC': 'Greece',
    'HRV': 'Croatia',
    'HUN': 'Hungary',
    'IRL': 'Ireland',
    'ISL': 'Iceland',
    'ITA': 'Italy',
    'LIE': 'Liechtenstein',
    'LTU': 'Lithuania',
    'LUX': 'Luxembourg',
    'LVA': 'Latvia',
    'MLT': 'Malta',
    'NLD': 'Netherlands',
    'NOR': 'Norway',
    'POL': 'Poland',
    'PRT': 'Portugal',
    'ROU': 'Romania',
    'SVK': 'Slovakia',
    'SVN': 'Slovenia',
    'SWE': 'Sweden'
}

INFRASTRUCTURE_TYPES = ["healthcare", "education", "power", "rail", "roads", "telecom"]
```

```python
# Base path for earthquake data
earthquake_path = Path("C:/Users/eks510/OneDrive - Vrije Universiteit Amsterdam/2_Projects/MIRACA/MIRACA_results")

# Initialize dictionary to store asset counts
asset_counts = {}

# Process each infrastructure type
for infra_type in INFRASTRUCTURE_TYPES:
    country_counts = {}
    
    # Process each country
    for country in COUNTRIES:
        file_path = earthquake_path / f"{country}_{infra_type}_earthquake_risk.parquet"
        
        if file_path.exists():
            # Read the parquet file
            df = pd.read_parquet(file_path)
            # Count the number of assets (rows)
            country_counts[country] = len(df)
        else:
            # No file exists for this country-infrastructure combination
            country_counts[country] = 0
    
    # Store counts for this infrastructure type
    asset_counts[infra_type] = country_counts

# Convert to DataFrame for easy viewing
asset_count_df = pd.DataFrame(asset_counts)

# Reindex to ensure all countries are included in correct order (already alphabetical)
asset_count_df = asset_count_df.reindex(COUNTRIES, fill_value=0)

# Add full country names as index
asset_count_df.index = [f"{code} - {COUNTRY_NAMES[code]}" for code in asset_count_df.index]

# Add a total row
total_row = asset_count_df.sum()
asset_count_df.loc['TOTAL'] = total_row

# Add a total column
asset_count_df['TOTAL'] = asset_count_df.sum(axis=1)

# Display the table
print("Infrastructure Asset Counts by Country and Type")
print("=" * 60)
print(asset_count_df)

# Save to CSV for reference
asset_count_df.to_csv('infrastructure_asset_counts.csv')
print(f"\nTable saved as 'infrastructure_asset_counts.csv'")

# Display some summary statistics
print(f"\nSummary Statistics:")
print(f"Total assets across all countries and infrastructure: {asset_count_df.loc['TOTAL', 'TOTAL']:,}")
print(f"Countries with data: {(asset_count_df.drop('TOTAL').sum(axis=1) > 0).sum()}")
print(f"Infrastructure types with data: {(asset_count_df.drop('TOTAL', axis=1).sum() > 0).sum()}")

asset_count_df
```
