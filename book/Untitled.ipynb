```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors

# Create the summary table based on the data you provided
data = {
    'Hazard Type': ['Earthquake', 'Flood', 'Landslide', 'Wind', 'Multi-hazard (E+F)'],
    'Roads': [11, 13, 45, 2, 0],
    'Bridges': [212, 0, 0, 0, 16],
    'Tunnels': [36, 0, 0, 0, 0],
    'Railway': [55, 8, 18, 0, 0]
}
summary_df = pd.DataFrame(data)

# Prepare data for heatmap
plot_data = summary_df.set_index('Hazard Type')

# Set style
sns.set_style("white")
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'Helvetica']

# Create the heatmap with green color scheme
fig, ax = plt.subplots(figsize=(11, 6))

# Custom green color palette based on your image
colors = ['#E0FDF2', '#CDFCEB', '#B5FAE1', '#90CF8D', '#64F400', '#5ADBB1', '#56CEA9', '#4CB499', '#429787']
cmap = mcolors.LinearSegmentedColormap.from_list('custom_green', colors)

sns.heatmap(plot_data, 
            annot=True, 
            fmt='d', 
            cmap=cmap,
            linewidths=1.5, 
            linecolor='white',
            #legend=False,
            cbar=False,  # This removes the colorbar

            #cbar_kws={'label': 'Number of curves', 'shrink': 0.8},
            square=False,
            annot_kws={'fontsize': 13, 'fontweight': 'bold'},
            vmin=0,
            ax=ax)

# Improved title
# ax.set_title('Distribution of Vulnerability and Fragility Curves\nby Hazard Type and Infrastructure', 
#              fontsize=16, fontweight='bold', pad=25, loc='left')

ax.set_xlabel(' ', fontsize=13, fontweight='bold', labelpad=10)
ax.set_ylabel(' ', fontsize=13, fontweight='bold', labelpad=10)

# Improve tick labels
plt.xticks(rotation=0, ha='center', fontsize=13, fontweight='600')
plt.yticks(rotation=0, fontsize=13, fontweight='600')

# Add a subtle background color
ax.set_facecolor('#f8f9fa')
fig.patch.set_facecolor('white')

# Remove spines for cleaner look
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.savefig('vulnerability_curves_heatmap.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
