```python
# Define infrastructure types
INFRASTRUCTURE_TYPES = ["healthcare", "education", "power", "rail", "roads", "telecom"]

base_path = Path("C:/Users/eks510/OneDrive - Vrije Universiteit Amsterdam/2_Projects/MIRACA/MIRACA_results/")

# Dynamically determine countries from coastal files in the directory
coastal_files = list(base_path.glob("*_coastal_*.parquet")) + list(base_path.glob("*_coastal_risk.parquet")) + \
                list(base_path.glob("*_coastal_*.csv")) + list(base_path.glob("*_coastal_risk.csv"))
countries_in_files = set()

for file in coastal_files:
    # Extract country code from filename (first part before underscore)
    country_code = file.stem.split('_')[0]
    countries_in_files.add(country_code)

COUNTRIES = sorted(list(countries_in_files))
print(f"Countries found in coastal files: {COUNTRIES}")
print(f"Total countries: {len(COUNTRIES)}")

# Helper function to read file (try parquet first, then CSV)
def read_coastal_file(file_path):
    """Read coastal file, trying parquet first, then CSV"""
    parquet_path = file_path.with_suffix('.parquet')
    csv_path = file_path.with_suffix('.csv')
    
    if parquet_path.exists():
        return pd.read_parquet(parquet_path)
    elif csv_path.exists():
        return pd.read_csv(csv_path)
    else:
        return None

base_path = Path("C:/Users/eks510/OneDrive - Vrije Universiteit Amsterdam/2_Projects/MIRACA/MIRACA_results/")
```

```python
# Initialize dictionaries to store data for each SSP scenario
coastal_data_ssp245 = {}
coastal_data_ssp585 = {}
INFRASTRUCTURE_TYPES = ["roads"]#["healthcare", "education", "power", "rail", "roads", "telecom"]

# Process each infrastructure type
for infra_type in INFRASTRUCTURE_TYPES:
    print(f"Processing {infra_type}...")
    
    # Initialize data containers
    infra_data_245 = []
    infra_data_585 = []
    
    for country in COUNTRIES:

        if country == 'BGR':
            continue
        # Baseline file (no SSP/year in filename)
        baseline_file = base_path / f"{country}_{infra_type}_coastal_risk"
        
        # SSP245 files
        file_2050_245 = base_path / f"{country}_{infra_type}_coastal_2050_SSP245_risk"
        file_2100_245 = base_path / f"{country}_{infra_type}_coastal_2100_SSP245_risk"
        
        # SSP585 files
        file_2050_585 = base_path / f"{country}_{infra_type}_coastal_2050_SSP585_risk"
        file_2100_585 = base_path / f"{country}_{infra_type}_coastal_2100_SSP585_risk"
        
        # Process SSP245 data (baseline + 2050 + 2100)
        country_data_245 = []
        
        df = read_coastal_file(baseline_file)
        if df is not None:
            df.loc[df.object_type.isin(['motorway','motorway_link', 'trunk', 'trunk_link'])]
            df['country'] = country
            df['year'] = 'Baseline'
            country_data_245.append(df)
        
        df = read_coastal_file(file_2050_245)
        if df is not None:
            df.loc[df.object_type.isin(['motorway','motorway_link', 'trunk', 'trunk_link'])]
            df['country'] = country
            df['year'] = '2050'
            country_data_245.append(df)
            
        df = read_coastal_file(file_2100_245)
        if df is not None:
            df.loc[df.object_type.isin(['motorway','motorway_link', 'trunk', 'trunk_link'])]
            df['country'] = country
            df['year'] = '2100'
            country_data_245.append(df)
        
        if country_data_245:
            infra_data_245.extend(country_data_245)
        
        # Process SSP585 data (baseline + 2050 + 2100)
        country_data_585 = []
        
        df = read_coastal_file(baseline_file)
        if df is not None:
            df.loc[df.object_type.isin(['motorway','motorway_link', 'trunk', 'trunk_link'])]
            df['country'] = country
            df['year'] = 'Baseline'
            country_data_585.append(df)
        
        df = read_coastal_file(file_2050_585)
        if df is not None:
            df.loc[df.object_type.isin(['motorway','motorway_link', 'trunk', 'trunk_link'])]
            df['country'] = country
            df['year'] = '2050'
            country_data_585.append(df)
            
        df = read_coastal_file(file_2100_585)
        if df is not None:
            df.loc[df.object_type.isin(['motorway','motorway_link', 'trunk', 'trunk_link'])]
            df['country'] = country
            df['year'] = '2100'
            country_data_585.append(df)
        
        if country_data_585:
            infra_data_585.extend(country_data_585)
    
    # Process SSP245 data
    if infra_data_245:
        infra_df_245 = pd.concat(infra_data_245, ignore_index=True)
        
        # Calculate country averages for each year (keep absolute EAD values)
        country_averages_245 = infra_df_245.groupby(['country', 'year'])['EAD'].sum().unstack(fill_value=0)
        
        # Ensure all time periods are present
        for period in ['Baseline', '2050', '2100']:
            if period not in country_averages_245.columns:
                country_averages_245[period] = 0
        
        # Reorder columns
        country_averages_245 = country_averages_245[['Baseline', '2050', '2100']]
        coastal_data_ssp245[infra_type] = country_averages_245
    
    # Process SSP585 data
    if infra_data_585:
        infra_df_585 = pd.concat(infra_data_585, ignore_index=True)
        
        # Calculate country averages for each year (keep absolute EAD values)
        country_averages_585 = infra_df_585.groupby(['country', 'year'])['EAD'].sum().unstack(fill_value=0)
        
        # Ensure all time periods are present
        for period in ['Baseline', '2050', '2100']:
            if period not in country_averages_585.columns:
                country_averages_585[period] = 0
        
        # Reorder columns
        country_averages_585 = country_averages_585[['Baseline', '2050', '2100']]
        coastal_data_ssp585[infra_type] = country_averages_585

print(f"SSP245 infrastructure types: {list(coastal_data_ssp245.keys())}")
print(f"SSP585 infrastructure types: {list(coastal_data_ssp585.keys())}")
```

```python
# Function to create the multi-panel plot
def create_coastal_plot(data_dict, scenario_name, countries_ordered=None):
    """Create multi-panel plot for coastal flooding EAD"""
    
    # If no country order provided, determine from data and filter out zero-only countries
    if countries_ordered is None:
        countries_with_data = set()
        for infra_type in INFRASTRUCTURE_TYPES:
            if infra_type in data_dict:
                # Only include countries that have actual data (not all zeros)
                countries_with_data_infra = data_dict[infra_type].loc[data_dict[infra_type].sum(axis=1) > 0].index
                countries_with_data.update(countries_with_data_infra)
        countries_ordered = sorted(list(countries_with_data))
    
    # Filter countries_ordered to only include those with actual data in this scenario
    countries_with_any_data = set()
    for infra_type in INFRASTRUCTURE_TYPES:
        if infra_type in data_dict:
            countries_with_data_infra = data_dict[infra_type].loc[data_dict[infra_type].sum(axis=1) > 0].index
            countries_with_any_data.update(countries_with_data_infra)
    
    # Keep only countries that have data for this specific scenario
    countries_ordered = [c for c in countries_ordered if c in countries_with_any_data]
    
    # Set style
    sns.set_style("whitegrid")
    
    # Create subplots
    fig, axes = plt.subplots(1, 6, figsize=(24, 16), sharey=True)
    
    # Define colors: Grey for baseline, Blue for 2050, Red for 2100
    colors = ['#808080', '#3498db', '#e74c3c']
    
    # Store legend info
    legend_handles = None
    legend_labels = None
    
    for i, infra_type in enumerate(INFRASTRUCTURE_TYPES):
        if infra_type in data_dict:
            # Get data and reindex to consistent country order
            df = data_dict[infra_type]
            # Only include countries that are in our filtered list
            df = df.reindex(countries_ordered, fill_value=0)
            
            # Create horizontal bar plot
            df.plot(kind='barh', ax=axes[i], width=0.8, color=colors)
            
            # Set log scale for x-axis
            axes[i].set_xscale('log')
            
            axes[i].set_title(f'{infra_type.capitalize()}', fontsize=28, fontweight='bold')
            axes[i].set_xlabel('EAD - Log Scale', fontsize=22)
            axes[i].tick_params(axis='both', which='major', labelsize=20)
            
            # Only show y-label for first subplot
            if i == 0:
                axes[i].set_ylabel('Country', fontsize=22)
            else:
                axes[i].set_ylabel('')
            
            # Store legend info from first plot with data
            if legend_handles is None:
                legend_handles, legend_labels = axes[i].get_legend_handles_labels()
            
            # Remove individual legends
            axes[i].legend().remove()
        
        else:
            # No data for this infrastructure type
            axes[i].text(0.5, 0.5, 'No Data', ha='center', va='center', 
                        transform=axes[i].transAxes, fontsize=22)
            axes[i].set_title(f'{infra_type.capitalize()}', fontsize=28, fontweight='bold')
    
    # Add horizontal legend above the plots
    if legend_handles:
        fig.legend(legend_handles, legend_labels, loc='upper center', 
                   bbox_to_anchor=(0.5, 0.92), ncol=len(legend_labels), 
                   fontsize=20, frameon=False)
    
    # plt.suptitle(f'Coastal Flooding Expected Annual Damage - {scenario_name}', 
    #              fontsize=30, fontweight='bold', y=0.99)
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    
    # Save the plot
    plt.savefig(f'coastal_flooding_EAD_{scenario_name.replace(" ", "_").replace(".", "")}.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# Determine consistent country ordering based on healthcare total EAD (small to large)
all_countries = set()
for scenario_data in [coastal_data_ssp245, coastal_data_ssp585]:
    for infra_type in INFRASTRUCTURE_TYPES:
        if infra_type in scenario_data:
            # Only include countries that have actual data (not all zeros)
            countries_with_data = scenario_data[infra_type].loc[scenario_data[infra_type].sum(axis=1) > 0].index
            all_countries.update(countries_with_data)

print(f"Countries with coastal data: {sorted(list(all_countries))}")

# Sort based on healthcare total EAD for SSP245 (baseline + future)
if 'healthcare' in coastal_data_ssp245:
    healthcare_df = coastal_data_ssp245['healthcare']
    # Only include countries with actual healthcare data
    healthcare_with_data = healthcare_df.loc[healthcare_df.sum(axis=1) > 0]
    healthcare_totals = healthcare_with_data.sum(axis=1)
    countries_ordered = healthcare_totals.sort_values(ascending=True).index.tolist()
    
    # Add any remaining countries with data but not in healthcare
    remaining_countries = sorted([c for c in all_countries if c not in countries_ordered])
    countries_ordered.extend(remaining_countries)
else:
    # Fallback: use countries with data from any infrastructure type
    countries_ordered = sorted(list(all_countries))

# Create plots for both scenarios
create_coastal_plot(coastal_data_ssp245, "SSP245", countries_ordered)
create_coastal_plot(coastal_data_ssp585, "SSP585", countries_ordered)
```

```python
# Initialize dictionary to store river data
river_data = {}

# Process each infrastructure type
for infra_type in INFRASTRUCTURE_TYPES:
    print(f"Processing {infra_type}...")
    
    # Initialize data container
    infra_data = []
    
    for country in COUNTRIES:
        # River climate file
        climate_file = base_path / f"{country}_{infra_type}_river_climate_risk.parquet"
        
        if climate_file.exists():
            df = pd.read_parquet(climate_file)
            df = df.loc[df.object_type.isin(['motorway','motorway_link', 'trunk', 'trunk_link'])]
            df['country'] = country
            infra_data.append(df)
    
    # Process data if we have any
    if infra_data:
        infra_df = pd.concat(infra_data, ignore_index=True)
        
        # Calculate country averages for baseline and warming scenarios (keep absolute EAD values)
        country_averages = infra_df.groupby('country')[['original_EAD', 'EAD_2.0C', 'EAD_3.0C', 'EAD_4.0C']].sum()
        
        # Rename columns for clarity
        country_averages.columns = ['Baseline', '2.0°C', '3.0°C', '4.0°C']
        
        # Store in dictionary
        river_data[infra_type] = country_averages

print(f"River infrastructure types with data: {list(river_data.keys())}")
```

```python

# Function to create the river flooding plot
def create_river_plot(data_dict):
    """Create multi-panel plot for river flooding EAD by warming scenario"""
    
    # Determine countries with data and filter out zero-only countries
    countries_with_data = set()
    for infra_type in INFRASTRUCTURE_TYPES:
        if infra_type in data_dict:
            # Only include countries that have actual data (not all zeros)
            countries_with_data_infra = data_dict[infra_type].loc[data_dict[infra_type].sum(axis=1) > 0].index
            countries_with_data.update(countries_with_data_infra)
    
    countries_ordered = sorted(list(countries_with_data))
    
    # Sort based on healthcare total EAD (small to large)
    if 'healthcare' in data_dict:
        healthcare_df = data_dict['healthcare']
        # Only include countries with actual healthcare data
        healthcare_with_data = healthcare_df.loc[healthcare_df.sum(axis=1) > 0]
        if len(healthcare_with_data) > 0:
            healthcare_totals = healthcare_with_data.sum(axis=1)
            countries_ordered = healthcare_totals.sort_values(ascending=True).index.tolist()
            
            # Add any remaining countries with data but not in healthcare
            remaining_countries = sorted([c for c in countries_with_data if c not in countries_ordered])
            countries_ordered.extend(remaining_countries)
    
    print(f"Countries with river data: {countries_ordered}")
    
    # Set style
    sns.set_style("whitegrid")
    
    # Create subplots
    fig, axes = plt.subplots(1, 6, figsize=(24, 16), sharey=True)
    
    # Define colors: Grey for baseline, Orange, Red, Dark Red for increasing warming
    colors = ['#808080', '#ff9800', '#f44336', '#8b0000']
    
    # Store legend info
    legend_handles = None
    legend_labels = None
    
    for i, infra_type in enumerate(INFRASTRUCTURE_TYPES):
        if infra_type in data_dict:
            # Get data and reindex to consistent country order
            df = data_dict[infra_type]
            # Only include countries that are in our filtered list
            df = df.reindex(countries_ordered, fill_value=0)
            
            # Create horizontal bar plot
            df.plot(kind='barh', ax=axes[i], width=0.8, color=colors)
            
            # Set log scale for x-axis
            axes[i].set_xscale('log')
            
            axes[i].set_title(f'{infra_type.capitalize()}', fontsize=28, fontweight='bold')
            axes[i].set_xlabel('EAD - Log Scale', fontsize=22)
            axes[i].tick_params(axis='both', which='major', labelsize=20)
            
            # Only show y-label for first subplot
            if i == 0:
                axes[i].set_ylabel('Country', fontsize=22)
            else:
                axes[i].set_ylabel('')
            
            # Store legend info from first plot with data
            if legend_handles is None:
                legend_handles, legend_labels = axes[i].get_legend_handles_labels()
            
            # Remove individual legends
            axes[i].legend().remove()
        
        else:
            # No data for this infrastructure type
            axes[i].text(0.5, 0.5, 'No Data', ha='center', va='center', 
                        transform=axes[i].transAxes, fontsize=22)
            axes[i].set_title(f'{infra_type.capitalize()}', fontsize=28, fontweight='bold')
    
    # Add horizontal legend above the plots
    if legend_handles:
        fig.legend(legend_handles, legend_labels, loc='upper center', 
                   bbox_to_anchor=(0.5, 0.96), ncol=len(legend_labels), 
                   fontsize=20, frameon=False)
    
    # plt.suptitle('River Flooding Expected Annual Damage by Global Warming Scenario', 
    #              fontsize=30, fontweight='bold', y=0.99)
    plt.tight_layout()
    plt.subplots_adjust(top=0.87)
    
    # Save the plot
    plt.savefig('river_flooding_EAD_warming_scenarios.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# Create the plot
create_river_plot(river_data)
```
