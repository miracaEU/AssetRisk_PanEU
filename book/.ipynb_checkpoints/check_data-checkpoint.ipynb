```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines

# Corridor names and colors
corridor_legend_data = {
    'A': ('Baltic Sea – Adriatic Sea', '#0080C0'),      # Blue
    'B': ('North Sea - Baltic', '#E91E8C'),             # Magenta/Pink
    'C': ('Mediterranean', '#00A651'),                  # Green
    'E': ('Scandinavian–Mediterranean', '#FF69B4'),     # Light Pink
    'G': ('Atlantic', '#FFD700'),                       # Yellow/Gold
    'I': ('Rhine – Danube', '#00BFFF'),                 # Cyan/Light Blue
    'J': ('Black Sea – Baltic Sea – Aegean Sea', '#8B4789'),  # Purple
    'K': ('Western Balkans – Eastern Mediterranean', '#8B4513'), # Brown
    'L': ('North Sea - Alpine', '#228B22'),             # Dark Green
}

# Create figure with no axes
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)
ax.axis('off')

# Create legend with circles
handles = []
for letter, (name, color) in corridor_legend_data.items():
    label = f"{letter} - {name}"
    handles.append(mlines.Line2D([], [], color=color, marker='o', linestyle='None',
                                  markersize=10, label=label))

# Create legend
legend = ax.legend(handles=handles, loc='center', frameon=False, fontsize=10)

plt.tight_layout()
plt.savefig('corridor_legend.png', dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```

```python
results = pd.read_parquet(r"C:\MIRACA\risk\roads_river_TENT_risk.parquet")
merged_results = gpd.GeoDataFrame(results.merge(base_roads[['id','geometry']],left_on='id',right_on='id'))
```

```python
import numpy as np

# Calculate total EAD for each corridor under each scenario
scenarios = ['original_EAD', 'EAD_1.5C', 'EAD_2.0C', 'EAD_3.0C', 'EAD_4.0C']
scenario_labels = ['Current', '1.5°C', '2.0°C', '3.0°C', '4.0°C']

# Manual colors for each scenario
scenario_colors = ['#4a4a4a', '#fee08b', '#fdae61', '#f46d43', '#d73027']

# Group by corridor and sum EAD for each scenario
corridor_risk = {}
for _, row in merged_results.iterrows():
    corridors = row['CORRIDORS']
    individual_corridors = list(corridors)
    
    for scenario in scenarios:
        ead = row[scenario]
        share = ead / len(individual_corridors)
        
        for corridor in individual_corridors:
            if corridor not in corridor_risk:
                corridor_risk[corridor] = {s: 0 for s in scenarios}
            corridor_risk[corridor][scenario] += share

# Convert to arrays for plotting, using full corridor names
corridor_names = [corridor_legend_data[c][0] for c in sorted(corridor_risk.keys())]
scenario_data = {scenario: [corridor_risk[c][scenario] for c in sorted(corridor_risk.keys())] for scenario in scenarios}

# Create horizontal grouped bar chart
fig, ax = plt.subplots(figsize=(8, 6))

y = np.arange(len(corridor_names))
height = 0.15

for i, (scenario, label, color) in enumerate(zip(scenarios, scenario_labels, scenario_colors)):
    offset = height * (i - 2)
    ax.barh(y + offset, [x/1e6 for x in scenario_data[scenario]], height, label=label, color=color)

ax.set_ylabel(' ', fontsize=12)
ax.set_xlabel('Total EAD (million EUR)', fontsize=12)
ax.set_yticks(y)
ax.set_yticklabels(corridor_names)
ax.legend(title='Warming Scenario', loc='upper right')
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig('future_river_risk.png', dpi=300, bbox_inches='tight', transparent=True)
```

```python
results = pd.read_parquet(r"C:\MIRACA\risk\roads_river_TENT_risk.parquet")
merged_results = gpd.GeoDataFrame(results.merge(base_roads[['id','geometry']],left_on='id',right_on='id'))
merged_results['EAD'] = merged_results['original_EAD']
# Filter for EAD above 1000
merged_results_filtered = merged_results[merged_results['EAD'] >= 1000].copy()
# Create logarithmic bins (starting from 1000)
bins = [1000, 10000, 100000, 200000, float('inf')]
labels = ['1k-10k', '10k-100k', '100k-200k', '>200k']
merged_results_filtered['EAD_category'] = pd.cut(
    merged_results_filtered['EAD'], 
    bins=bins, 
    labels=labels, 
    include_lowest=True
)

# Expand corridor tags and sum damages by individual corridor
corridor_damage_dict = {}
for _, row in merged_results_filtered.iterrows():
    corridors = row['CORRIDORS']
    ead = row['EAD']
    # Split the corridor string into individual letters and distribute damage
    individual_corridors = list(corridors)
    share = ead / len(individual_corridors)  # Divide damage equally among corridors
    for corridor in individual_corridors:
        if corridor in corridor_damage_dict:
            corridor_damage_dict[corridor] += share
        else:
            corridor_damage_dict[corridor] = share

# Convert to Series and sort
corridor_damage = pd.Series(corridor_damage_dict).sort_values(ascending=False)

# Create the map
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
# Set style
plt.rcParams['font.family'] = 'sans-serif'
# Custom color palette (orange to red for high values)
colors = ['#bb3e03','#9b2226','#540b0e','#250902']
color_dict = {label: colors[i] for i, label in enumerate(labels)}
# Ensure both dataframes are in Web Mercator for contextily
base_data_merc = base_roads.to_crs(epsg=3857)
merged_results_filtered_merc = merged_results_filtered.to_crs(epsg=3857)
# Plot base_data as thin gray lines in the background
base_data_merc.plot(ax=ax, color='#cccccc', linewidth=1)
# Plot the roads colored by EAD category
for category in labels:
    subset = merged_results_filtered_merc[merged_results_filtered_merc['EAD_category'] == category]
    if len(subset) > 0:
        subset.plot(ax=ax, color=color_dict[category], linewidth=1.5, 
                   label=f'{category} EUR', alpha=0.9)
# Add contextily basemap
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron, alpha=0.6)

# Create inset axes for pie chart
ax_inset = fig.add_axes([0.22, 0.62, 0.25, 0.25])

# And for the pie chart dictionary:
corridor_color_map = {
    'A': '#0080C0',      # Baltic Sea – Adriatic Sea (Blue)
    'B': '#E91E8C',      # North Sea - Baltic (Magenta/Pink)
    'C': '#00A651',      # Mediterranean (Green)
    'E': '#FF69B4',      # Scandinavian–Mediterranean (Light Pink)
    'G': '#FFD700',      # Atlantic (Yellow/Gold)
    'I': '#00BFFF',      # Rhine – Danube (Cyan/Light Blue)
    'J': '#8B4789',      # Black Sea – Baltic Sea – Aegean Sea (Purple)
    'K': '#8B4513',      # Western Balkans – Eastern Mediterranean (Brown)
    'L': '#228B22',      # North Sea - Alpine (Dark Green)
    'Other': '#bc80bd'   # Other
}
# After grouping corridors, map colors based on corridor letter
corridor_colors_mapped = [corridor_color_map.get(corridor, '#cccccc') for corridor in corridor_damage.index]

# Then use in pie chart:
wedges, texts, autotexts = ax_inset.pie(
    corridor_damage.values, 
    labels=corridor_damage.index,
    autopct='%1.1f%%',
    startangle=180,
    textprops={'fontsize': 8},
    colors=corridor_colors_mapped
)

# Style the percentage text
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(7)

# Manually set limits
ax.set_ylim(4200000, 8900000)
ax.set_xlim(-1100000, 3700000)
# Remove axis ticks
ax.set_xticks([])
ax.set_yticks([])
ax.set_axis_off()

plt.savefig('TENT_road_river.png',dpi=300,bbox_inches='tight')
```

```python
results = pd.read_parquet(r"C:\MIRACA\risk\rail_coastal_TENT_risk.parquet")
merged_results = gpd.GeoDataFrame(results.merge(base_rail[['id','geometry']],left_on='id',right_on='id'))
# Filter for EAD above 1000
merged_results_filtered = merged_results[merged_results['EAD'] >= 1000].copy()
# Create logarithmic bins (starting from 1000)
bins = [1000, 10000, 100000, 200000, float('inf')]
labels = ['1k-10k', '10k-100k', '100k-200k', '>200k']
merged_results_filtered['EAD_category'] = pd.cut(
    merged_results_filtered['EAD'], 
    bins=bins, 
    labels=labels, 
    include_lowest=True
)

# Expand corridor tags and sum damages by individual corridor
corridor_damage_dict = {}
for _, row in merged_results_filtered.iterrows():
    corridors = row['CORRIDORS']
    ead = row['EAD']
    # Split the corridor string into individual letters and distribute damage
    individual_corridors = list(corridors)
    share = ead / len(individual_corridors)  # Divide damage equally among corridors
    for corridor in individual_corridors:
        if corridor in corridor_damage_dict:
            corridor_damage_dict[corridor] += share
        else:
            corridor_damage_dict[corridor] = share

# Convert to Series and sort
corridor_damage = pd.Series(corridor_damage_dict).sort_values(ascending=False)

# Create the map
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
# Set style
plt.rcParams['font.family'] = 'sans-serif'
# Custom color palette (orange to red for high values)
colors = ['#bb3e03','#9b2226','#540b0e','#250902']
color_dict = {label: colors[i] for i, label in enumerate(labels)}

# Ensure both dataframes are in Web Mercator for contextily
base_data_merc = base_rail.to_crs(epsg=3857)
merged_results_filtered_merc = merged_results_filtered.to_crs(epsg=3857)
# Plot base_data as thin gray lines in the background
base_data_merc.plot(ax=ax, color='#cccccc', linewidth=1)
# Plot the roads colored by EAD category
for category in labels:
    subset = merged_results_filtered_merc[merged_results_filtered_merc['EAD_category'] == category]
    if len(subset) > 0:
        subset.plot(ax=ax, color=color_dict[category], linewidth=1.5, 
                   label=f'{category} EUR', alpha=0.9)
# Add contextily basemap
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron, alpha=0.6)

# Create inset axes for pie chart
ax_inset = fig.add_axes([0.22, 0.62, 0.25, 0.25])

# And for the pie chart dictionary:
corridor_color_map = {
    'A': '#0080C0',      # Baltic Sea – Adriatic Sea (Blue)
    'B': '#E91E8C',      # North Sea - Baltic (Magenta/Pink)
    'C': '#00A651',      # Mediterranean (Green)
    'E': '#FF69B4',      # Scandinavian–Mediterranean (Light Pink)
    'G': '#FFD700',      # Atlantic (Yellow/Gold)
    'I': '#00BFFF',      # Rhine – Danube (Cyan/Light Blue)
    'J': '#8B4789',      # Black Sea – Baltic Sea – Aegean Sea (Purple)
    'K': '#8B4513',      # Western Balkans – Eastern Mediterranean (Brown)
    'L': '#228B22',      # North Sea - Alpine (Dark Green)
    'Other': '#bc80bd'   # Other
}

# Then modify the pie chart call to use these colors:
wedges, texts, autotexts = ax_inset.pie(
    corridor_damage.values, 
    labels=corridor_damage.index,
    autopct='%1.1f%%',
    startangle=180,
    textprops={'fontsize': 8, 'fontweight': 'bold'},
    colors=corridor_colors_mapped
)

# Style the percentage text
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(7)

# Manually set limits
ax.set_ylim(4200000, 8900000)
ax.set_xlim(-1100000, 3700000)
# Remove axis ticks
ax.set_xticks([])
ax.set_yticks([])
ax.set_axis_off()

plt.savefig('TENT_rail_coastal.png',dpi=300,bbox_inches='tight')
```
