# Test: MultiLineString fix for Spain rail

Reproduces and verifies the fix for:
> `TypeError: 'MultiLineString' object is not subscriptable`

which occurs in `damagescanner._convert_to_meters` when features contain `MultiLineString` geometries.

```python
import sys
sys.path.insert(0, '../src')  # adjust if your src path differs

import numpy as np
import pandas as pd
import geopandas as gpd
import xarray as xr
import matplotlib.pyplot as plt
from pathlib import Path
from shapely.geometry import MultiLineString, LineString
```

## 1. Load Spain rail exposure data

```python
EXPOSURE_DIR = Path(
    r"C:\Users\eks510\OneDrive - Vrije Universiteit Amsterdam"
    r"\Documenten - MIRACA\General\Deliverables and Milestones"
    r"\Milestone_Intermediate version of harmonised exposure database"
    r"\Exposure_database_new_version_2026\Exposure_files"
)

rail_path = EXPOSURE_DIR / "Railway" / "Railway_ES.parquet"
print(f"Loading: {rail_path}")
print(f"Exists:  {rail_path.exists()}")

features = gpd.read_parquet(rail_path)
print(f"\nLoaded {len(features)} features")
features.head()
```

## 2. Inspect geometry types

```python
geom_counts = features.geometry.geom_type.value_counts()
print("Geometry type breakdown:")
print(geom_counts)

n_multi = (features.geometry.geom_type == "MultiLineString").sum()
print(f"\nMultiLineString count: {n_multi} ({n_multi/len(features)*100:.1f}%)")
```

```python
# Plot a sample to visualise
fig, ax = plt.subplots(figsize=(10, 8))

single = features[features.geometry.geom_type == "LineString"]
multi  = features[features.geometry.geom_type == "MultiLineString"]

if len(single): single.plot(ax=ax, color="steelblue", linewidth=0.5, label=f"LineString ({len(single)})")
if len(multi):  multi.plot(ax=ax,  color="red",       linewidth=1.0, label=f"MultiLineString ({len(multi)})")

ax.legend()
ax.set_title("Spain rail — geometry types")
plt.tight_layout()
plt.show()
```

## 3. Reproduce the crash (before fix)

This cell calls `damagescanner._convert_to_meters` directly on a MultiLineString row to confirm the bug.

```python
from damagescanner.vector import _convert_to_meters

multi_rows = features[features.geometry.geom_type == "MultiLineString"]

if len(multi_rows) == 0:
    print("No MultiLineString rows found — bug would not be triggered for this file.")
else:
    sample = multi_rows.iloc[0]
    print(f"Sample MultiLineString: {sample.geometry}")
    print("\nCalling _convert_to_meters (expect crash without fix)...")
    try:
        result = _convert_to_meters(sample)
        print(f"  Result: {result}  ← no crash (damagescanner may have been patched)")
    except TypeError as e:
        print(f"  ✗ Crashed as expected: {e}")
        print("    → This confirms the bug. Fix is in risk_integration.explode_multi()")
```

## 4. Test the fix: explode_multi + reaggregate_damage

```python
from risk_integration import explode_multi, reaggregate_damage

exploded = explode_multi(features)

print(f"Original:  {len(features)} features")
print(f"Exploded:  {len(exploded)} features (+{len(exploded)-len(features)} parts)")
print(f"\nGeometry types after explode:")
print(exploded.geometry.geom_type.value_counts())

assert "_orig_idx" in exploded.columns, "Missing _orig_idx column"
assert (exploded.geometry.geom_type == "MultiLineString").sum() == 0, "MultiLineStrings still present!"
print("\n✓ No MultiLineStrings remain after explode")
```

```python
# Simulate damage results on exploded features and re-aggregate
fake_damage = exploded.copy()
fake_damage["damage_mean"] = np.random.uniform(0, 10000, len(fake_damage))
fake_damage["damage_min"]  = fake_damage["damage_mean"] * 0.5
fake_damage["damage_max"]  = fake_damage["damage_mean"] * 2.0

reagg = reaggregate_damage(fake_damage, features)

print(f"Re-aggregated: {len(reagg)} features (expect {len(features)})")
assert len(reagg) == len(features), "Row count mismatch after reaggregation!"
print("\n✓ reaggregate_damage restored original feature count")

# Verify multi-part features have summed damage
multi_idx = features[features.geometry.geom_type == "MultiLineString"].index
if len(multi_idx) > 0:
    parts = exploded[exploded["_orig_idx"] == multi_idx[0]]
    expected_sum = parts["damage_mean"].sum()
    actual = reagg.loc[multi_idx[0], "damage_mean"]
    print(f"\nSample multi-part feature ({multi_idx[0]}):")
    print(f"  Parts sum:      {expected_sum:.2f}")
    print(f"  Re-agg result:  {actual:.2f}")
    assert abs(expected_sum - actual) < 0.01, "Damage sum mismatch!"
    print("  ✓ Damage correctly summed across parts")
```

## 5. End-to-end test: run VectorScanner on exploded features

Load one river hazard raster and run a single RP through `compute_damage_per_rp` to confirm the full path works.

```python
RIVER_HAZARD_DIR = Path(
    r"C:\Users\eks510\OneDrive - Vrije Universiteit Amsterdam"
    r"\Documenten - MIRACA\WP3\D3.2\Hazard_data\River_floods"
)
VULNERABILITY_PATH = Path(r"Table_D2_Hazard_Fragility_and_Vulnerability_Curves_V1.1.0_conversions.xlsx")

# Load RP100 only for a quick test
raster_path = RIVER_HAZARD_DIR / "Europe_RP100_filled_depth.tif"
print(f"Hazard raster: {raster_path}")
print(f"Exists: {raster_path.exists()}")
```

```python
from hazard_river import prepare_flood_curves, get_country_bounds_4326
from risk_integration import compute_damage_per_rp

# Clip raster to Spain extent
bounds = get_country_bounds_4326(features)
hazard = xr.open_dataset(raster_path, engine="rasterio")
hazard = hazard.rio.clip_box(*bounds)
print(f"Hazard raster clipped to Spain: {hazard.dims}")

# Load curves
damage_curves, multi_curves, maxdam, _, _ = prepare_flood_curves("rail", VULNERABILITY_PATH)
print(f"Curves loaded: {list(multi_curves.keys())}")
```

```python
# Run compute_damage_per_rp — this internally calls explode_multi
print("Running compute_damage_per_rp on Spain rail (RP100)...")

result = compute_damage_per_rp(
    features=features,
    hazard=hazard,
    curve_path=damage_curves,
    maxdam=maxdam,
    asset_type="rail",
    multi_curves=multi_curves,
)

print(f"\n✓ Success! Result shape: {result.shape}")
print(f"  Features with damage > 0: {(result['damage_mean'] > 0).sum()}")
print(f"  Total damage (mean):      {result['damage_mean'].sum():.2e}")
result[["object_type", "damage_mean", "damage_min", "damage_max"]].describe()
```

```python
# Plot damage distribution
damaged = result[result["damage_mean"] > 0]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Map
result.plot(column="damage_mean", ax=axes[0], legend=True,
            cmap="YlOrRd", missing_kwds={"color": "lightgrey"})
axes[0].set_title(f"Spain rail — damage at RP100\n({len(damaged)} assets affected)")

# Histogram
if len(damaged) > 0:
    damaged["damage_mean"].clip(upper=damaged["damage_mean"].quantile(0.99)).hist(
        ax=axes[1], bins=40, color="steelblue", edgecolor="white"
    )
    axes[1].set_xlabel("Damage (€)")
    axes[1].set_ylabel("Count")
    axes[1].set_title("Damage distribution (99th pctile cap)")

plt.tight_layout()
plt.show()

print("\n✓ MultiLineString fix confirmed — full pipeline runs without error.")
```
