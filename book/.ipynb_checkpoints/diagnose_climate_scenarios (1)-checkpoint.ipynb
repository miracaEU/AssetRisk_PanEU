```python
import sys
sys.path.insert(0, '../src')
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from pathlib import Path

BASIN_PATH   = Path(r'C:\Users\eks510\OneDrive - Vrije Universiteit Amsterdam\12_repositories\AssetRisk_PanEU\src\basins_abs_shift_return_periods.parquet')
EXPOSURE_DIR = Path(r'C:\Users\eks510\OneDrive - Vrije Universiteit Amsterdam\Documenten - MIRACA\General\Deliverables and Milestones\Milestone_Intermediate version of harmonised exposure database\Exposure_database_new_version_2026\Exposure_files')
```

## 1. Basin data structure

```python
basin = gpd.read_parquet(BASIN_PATH)
print(f'Shape: {basin.shape}')
print(f'Index name: {basin.index.name}')
print(f'Index sample: {basin.index[:5].tolist()}')
print(f'Columns: {basin.columns.tolist()}')
```

## 2. Assign GRC rail features to basins spatially

```python
from hazard_river import assign_basin_ids

features  = gpd.read_parquet(EXPOSURE_DIR / 'Railway' / 'Railway_EL.parquet')
basin_ids = assign_basin_ids(features, basin)

matched_ids = basin_ids.dropna().unique()
print(f'Features: {len(features)}')
print(f'Matched to a basin: {basin_ids.notna().sum()} ({basin_ids.notna().mean()*100:.1f}%)')
print(f'Unique basins: {len(matched_ids)}')
```

## 3. Check monotonicity of RP shifts across warming levels

```python
temps  = ['15', '20', '30', '40']
labels = ['1.5C', '2.0C', '3.0C', '4.0C']
valid_ids = [b for b in matched_ids if b in basin.index]

print('Non-monotonic basin counts (where higher warming => lower RP shift):')
for rp in [10, 100, 500]:
    cols = [f'{rp}_rp_change_{t}' for t in temps]
    d = basin.loc[valid_ids, cols].copy()
    d.columns = labels
    for a, b in [('1.5C','2.0C'), ('2.0C','3.0C'), ('3.0C','4.0C')]:
        n = (d[b] < d[a]).sum()
        print(f'  RP{rp:4d}  {a} -> {b}: {n}/{len(d)} non-monotonic ({n/len(d)*100:.1f}%)')
```

```python
# Distribution of RP100 shifts
cols = [f'100_rp_change_{t}' for t in temps]
shift100 = basin.loc[valid_ids, cols].copy()
shift100.columns = labels

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

shift100.boxplot(ax=axes[0])
axes[0].axhline(100, color='red', linestyle='--', label='baseline RP100')
axes[0].set_title('RP100 shift per scenario — GRC basins')
axes[0].set_ylabel('New return period (years)')
axes[0].legend()

axes[1].scatter(shift100['3.0C'], shift100['4.0C'], alpha=0.5, s=10)
lim = shift100[['3.0C','4.0C']].max().max()
axes[1].plot([100, lim], [100, lim], 'r--', label='1:1 line (monotonic)')
axes[1].set_xlabel('RP100 shift at 3.0C')
axes[1].set_ylabel('RP100 shift at 4.0C')
axes[1].set_title('3.0C vs 4.0C — points below line = non-monotonic')
axes[1].legend()
plt.tight_layout()
plt.show()
```

## 4. Trace the EAD computation for a sample basin

```python
from risk_integration import adjust_return_periods_climate

# Pick the basin covering the most features
dominant_basin = basin_ids.value_counts().index[0]
basin_row = basin.loc[dominant_basin]
base_rps  = [10, 20, 30, 40, 50, 75, 100, 200, 500]

print(f'Dominant basin: {dominant_basin} ({basin_ids.value_counts().iloc[0]} features)')
print(f'\nRP shifts for base RPs {base_rps}:')
print(f'{"":>6} ' + '  '.join(f'{r:>6}' for r in base_rps))
for temp, label in zip(temps, labels):
    adj, _ = adjust_return_periods_climate(base_rps, 0, basin_row, temp)
    print(f'{label:>6} ' + '  '.join(f'{a:>6.0f}' for a in adj))
```

```python
# Show how the RP shift affects the exceedance probability curve
# Higher RP = lower annual probability = lower EAD
# So if 4.0C has higher RPs than 3.0C, EAD at 4.0C < EAD at 3.0C

fig, ax = plt.subplots(figsize=(10, 5))
colors = ['#2196F3', '#4CAF50', '#FF9800', '#F44336']

for (temp, label), color in zip(zip(temps, labels), colors):
    adj, _ = adjust_return_periods_climate(base_rps, 0, basin_row, temp)
    probs  = [1/r for r in adj]
    ax.plot(base_rps, probs, marker='o', label=label, color=color)

base_probs = [1/r for r in base_rps]
ax.plot(base_rps, base_probs, 'k--', marker='x', label='baseline', zorder=5)
ax.set_xlabel('Base return period (years)')
ax.set_ylabel('Annual exceedance probability (adjusted)')
ax.set_title(f'Exceedance probability curve per scenario — basin {dominant_basin}\n(lower = less frequent = lower EAD contribution)')
ax.legend()
plt.tight_layout()
plt.show()

print('\nIf 4.0C curve is BELOW 3.0C at high RPs, that explains lower EAD at 4.0C')
```
