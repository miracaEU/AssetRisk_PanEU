# Diagnose: Climate scenario EADs identical to baseline
Checks basin data alignment with exposure features.

```python
import sys
sys.path.insert(0, '../src')

import pandas as pd
import geopandas as gpd
from pathlib import Path

BASIN_PATH   = Path(r"C:\Users\eks510\OneDrive - Vrije Universiteit Amsterdam\12_repositories\AssetRisk_PanEU\src\basins_abs_shift_return_periods.parquet")
EXPOSURE_DIR = Path(r"C:\Users\eks510\OneDrive - Vrije Universiteit Amsterdam\Documenten - MIRACA\General\Deliverables and Milestones\Milestone_Intermediate version of harmonised exposure database\Exposure_database_new_version_2026\Exposure_files")
```

## 1. Inspect basin data

```python
basin = pd.read_parquet(BASIN_PATH)
print(f"Shape: {basin.shape}")
print(f"Index name: {basin.index.name}")
print(f"Index sample: {basin.index[:10].tolist()}")
print(f"\nColumns: {basin.columns.tolist()}")
```

```python
# Check which column names the climate code expects
expected_cols = [
    '10_rp_change_15',  '100_rp_change_15',  '500_rp_change_15',
    '10_rp_change_20',  '100_rp_change_20',  '500_rp_change_20',
    '10_rp_change_30',  '100_rp_change_30',  '500_rp_change_30',
    '10_rp_change_40',  '100_rp_change_40',  '500_rp_change_40',
]
print("Expected columns present?")
for col in expected_cols:
    print(f"  {col}: {'✓' if col in basin.columns else '✗ MISSING'}")
```

## 2. Inspect GRC rail exposure

```python
features = gpd.read_parquet(EXPOSURE_DIR / 'Railway' / 'Railway_EL.parquet')
print(f"Features shape: {features.shape}")
print(f"Columns: {features.columns.tolist()}")
print(f"\nNUTS2 sample: {features['NUTS2'].dropna().unique()[:10]}")
print(f"NUTS2 unique count: {features['NUTS2'].nunique()}")
print(f"NUTS2 nulls: {features['NUTS2'].isna().sum()}")
```

## 3. Check join between NUTS2 and basin index

```python
feature_nuts2  = set(features['NUTS2'].dropna().unique())
basin_ids      = set(basin.index)

matched   = feature_nuts2 & basin_ids
unmatched = feature_nuts2 - basin_ids

print(f"Feature NUTS2 values:  {len(feature_nuts2)}")
print(f"Basin index values:    {len(basin_ids)}")
print(f"Matched:               {len(matched)}")
print(f"Unmatched (no basin):  {len(unmatched)}")

if unmatched:
    print(f"\nUnmatched NUTS2: {sorted(unmatched)}")
if matched:
    print(f"\nMatched sample:  {sorted(matched)[:5]}")
```

## 4. Check actual shift values for matched basins

```python
if matched and all(c in basin.columns for c in ['100_rp_change_15', '100_rp_change_40']):
    sample = basin.loc[list(matched)[:5], ['10_rp_change_15','100_rp_change_15','500_rp_change_15',
                                            '10_rp_change_40','100_rp_change_40','500_rp_change_40']]
    print("RP shift values for sample matched basins:")
    print(sample)
    print(f"\nAre all shift values == original RPs? (would mean no change)")
    print(f"  100_rp_change_15 == 100: {(basin.loc[list(matched), '100_rp_change_15'] == 100).all()}")
    print(f"  100_rp_change_40 == 100: {(basin.loc[list(matched), '100_rp_change_40'] == 100).all()}")
else:
    print("Cannot check — either no matches or column names differ.")
    print(f"\nActual basin columns: {basin.columns.tolist()}")
```

## 5. Trace adjust_return_periods_climate for one asset

```python
from risk_integration import adjust_return_periods_climate

if matched:
    sample_id  = list(matched)[0]
    basin_row  = basin.loc[sample_id]
    base_rps   = [10, 20, 30, 40, 50, 75, 100, 200, 500]

    print(f"Basin: {sample_id}")
    for temp in ('15', '20', '30', '40'):
        adj_rps, _ = adjust_return_periods_climate(base_rps, 0, basin_row, temp)
        changed = [f"{o}→{a:.0f}" for o, a in zip(base_rps, adj_rps) if abs(o-a) > 0.5]
        print(f"  {temp}°C: {changed if changed else 'NO CHANGE (all identical)'}")
else:
    print("No matched basins to test.")
```
